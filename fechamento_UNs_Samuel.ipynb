```python
import pandas as pd
import zipfile
import os
import glob
import requests
import shutil
pasta_origem = "//Srvl002/data/FINANCEIRO/PLANNING/INTELIGENCIA DE DADOS"
```

```python
# dataInicial e dataFinal formato: 'dd/mm/yyyy'
def consulta_bacen(codigo_bcb, dataInicial, dataFinal, column_name):
    if type(codigo_bcb) == list:
        url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=csv&dataInicial={}&dataFinal={}'.format(codigo_bcb[0], dataInicial, dataFinal)
        i = 1
        df = pd.read_csv(url, sep=';')
        tam_cod = len(codigo_bcb)
        if tam_cod != len(column_name):
            print('O tamanho da lista do código do bacen não é igual a quantidade de nome das colunas!')
            return
        while i < tam_cod:
            url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=csv&dataInicial={}&dataFinal={}'.format(codigo_bcb[i], dataInicial, dataFinal)
            df_aux = pd.read_csv(url, sep=';')
            df = df.merge(df_aux, how='outer', on='data', suffixes=(i-1, i))
            i += 1
        df_columns = ['Data'] + column_name
        df.columns = df_columns
        return df
    else:
        url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=csv&dataInicial={}&dataFinal={}'.format(codigo_bcb, dataInicial, dataFinal)
        df = pd.read_csv(url, sep=';')
        df_columns = ['Data'].append(column_name)
        df.columns = df_columns
        return df
```

```python
# folder path
dir_path = pasta_origem+'/18. Python Notebooks/2.Dados de Mercado/Consórcio/Arquivos/'

# list to store files
res = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res.append(path)
print(res)
```

```python
for mes_download in anomes:
    nome_arquivo = mes_download + 'Consorcios.zip'
    if nome_arquivo in res:
        next
    else:
        url = 'https://www.bcb.gov.br/Fis/Consorcios/Port/BD/' + mes_download + 'Consorcios.zip'
        filename = pasta_origem+'/18. Python Notebooks/2.Dados de Mercado/Consórcio/Arquivos/' + mes_download + 'Consorcios.zip'

        response = requests.get(url)
        if response.status_code != 200:
            next
        else:
            with open(filename, 'wb') as f:
                f.write(response.content)
            res.append(nome_arquivo)
```

```python
# folder path
dir_path = pasta_origem+'/18. Python Notebooks/2.Dados de Mercado/Consórcio/src/'

# list to store files
res_csv = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res_csv.append(path)
print(res_csv)
```

```python
for zip_file in res:
    with zipfile.ZipFile(pasta_origem+"/18. Python Notebooks/2.Dados de Mercado/Consórcio/Arquivos/" + zip_file, "r") as zip_ref:
        for filename in zip_ref.namelist():
            if filename.endswith('Consolidados.csv') and filename in res_csv:
                next
            else:
                if filename.endswith('Consolidados.csv'):
                    print(filename) 
                    zip_ref.extract(filename, path=pasta_origem+'/18. Python Notebooks/2.Dados de Mercado/Consórcio/src/')
```

```python
ano_mes = '202101|202102|202103|202104|202105|202106|202107|202108|202109|202110|202111|202112|202201|202202|202203|202204|202205|202206|202207|202208|202209|202210|202211|202212|202301|202302|202303|202304|202305|202306|202307|202308|202309|202310|202311|202312|202401|202402|202403|202404|202405|202406|202407|202408|202409|202410|202411|202412|202501|202502|202503|202504|202505|202506|202507|202508|202509|202510|202511|202512'
```

```python
link = 'https://servicodados.ibge.gov.br/api/v3/agregados/8882/periodos/' + ano_mes + '/variaveis/7169?localidades=N1[all]|N3[all]&classificacao=11046[all]|85[31556,103158]'

requisicao = requests.get(link)
informacoes = requisicao.json()
```

```python
df_receita_eletrodomestico = pd.json_normalize(informacoes[0]['resultados'][0]['series'])
df_receita_outros_artigos = pd.json_normalize(informacoes[0]['resultados'][1]['series'])
df_volume_eletrodomestico = pd.json_normalize(informacoes[0]['resultados'][2]['series'])
df_volume_outros_artigos = pd.json_normalize(informacoes[0]['resultados'][3]['series'])
```

```python
df_receita_eletrodomestico = df_receita_eletrodomestico.set_index(df_receita_eletrodomestico['localidade.nome']).drop(columns=['localidade.id', 'localidade.nivel.id', 'localidade.nivel.nome', 'localidade.nome'])
df_receita_eletrodomestico = df_receita_eletrodomestico.unstack().reset_index()
df_receita_eletrodomestico['Indice'] = 'Receita'
df_receita_eletrodomestico['Categoria'] = 'Eletrodoméstico'

df_receita_outros_artigos = df_receita_outros_artigos.set_index(df_receita_outros_artigos['localidade.nome']).drop(columns=['localidade.id', 'localidade.nivel.id', 'localidade.nivel.nome', 'localidade.nome'])
df_receita_outros_artigos = df_receita_outros_artigos.unstack().reset_index()
df_receita_outros_artigos['Indice'] = 'Receita'
df_receita_outros_artigos['Categoria'] = 'Outros artigos de uso pessoal e doméstico'

df_volume_eletrodomestico = df_volume_eletrodomestico.set_index(df_volume_eletrodomestico['localidade.nome']).drop(columns=['localidade.id', 'localidade.nivel.id', 'localidade.nivel.nome', 'localidade.nome'])
df_volume_eletrodomestico = df_volume_eletrodomestico.unstack().reset_index()
df_volume_eletrodomestico['Indice'] = 'Volume'
df_volume_eletrodomestico['Categoria'] = 'Eletrodoméstico'

df_volume_outros_artigos = df_volume_outros_artigos.set_index(df_volume_outros_artigos['localidade.nome']).drop(columns=['localidade.id', 'localidade.nivel.id', 'localidade.nivel.nome', 'localidade.nome'])
df_volume_outros_artigos = df_volume_outros_artigos.unstack().reset_index()
df_volume_outros_artigos['Indice'] = 'Volume'
df_volume_outros_artigos['Categoria'] = 'Outros artigos de uso pessoal e doméstico'
```

```python
df_final = pd.concat([df_receita_eletrodomestico, df_receita_outros_artigos, df_volume_eletrodomestico, df_volume_outros_artigos])
df_final.columns = ['anomes', 'localidade', 'valor', 'indice', 'categoria']
df_final['anomes'] = df_final['anomes'].str.replace('serie.', "")
df_final['valor'] = df_final['valor'].astype('float64')
df_final.to_csv(pasta_origem+'/18. Python Notebooks/2.Dados de Mercado/IBGE/consolidado_varejo.csv', sep=';', encoding='ISO-8859-1', decimal=',', index=False)
df_final.to_csv('C:/Users/samuelbarroso/Wiz Co/Ciclo & Inteligência - Documentos/Bases/11-Mercado/IBGE/consolidado_varejo.csv', sep=';', encoding='ISO-8859-1', decimal=',', index=False)
df_final
```

```python
# projeção selic
data = pd.read_csv('https://olinda.bcb.gov.br/olinda/servico/Expectativas/versao/v1/odata/ExpectativasMercadoAnuais?$format=text/csv&$select=Indicador,Data,DataReferencia,Media,Mediana,DesvioPadrao,Minimo,Maximo,numeroRespondentes,baseCalculo')
df = data[(data['Indicador'] == 'Selic') & (data['Data'] == data['Data'].max()) & (data['baseCalculo'] == 1)]
df['Anomes'] = df['DataReferencia'].astype('str')  + '12'
df['DataPrev'] = df['DataReferencia'].astype('str') + '-12-01'
```

```python
dados_bmg['Data'] = pd.to_datetime(dados_bmg['Data'], dayfirst=True)
```

```python
dados_bmg['Projecao Selic Mediana'] = None
dados_bmg['Projecao Selic Minima'] = None
dados_bmg['Projecao Selic Maxima'] = None
new_row = {'Data': dados_bmg['Data'].max().replace(day=1), 
           'Projecao Selic Mediana': dados_bmg[dados_bmg['Data'] == dados_bmg['Data'].max()]['Taxa de juros - Meta Selic definida pelo Copom'].values[0],
           'Projecao Selic Minima': dados_bmg[dados_bmg['Data'] == dados_bmg['Data'].max()]['Taxa de juros - Meta Selic definida pelo Copom'].values[0],
           'Projecao Selic Maxima': dados_bmg[dados_bmg['Data'] == dados_bmg['Data'].max()]['Taxa de juros - Meta Selic definida pelo Copom'].values[0]}
dados_bmg.loc[len(dados_bmg)] = new_row
```
