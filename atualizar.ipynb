```python
import os
import pandas as pd
import unicodedata

folder = r"C:\Users\samuelbarroso\Downloads\Agencias\Agencias"
output_fileXLS = r"C:\Users\samuelbarroso\Downloads\Agencias\Agencias_Consolidadoxls.csv"
output_fileXLSX = r"C:\Users\samuelbarroso\Downloads\Agencias\Agencias_Consolidadoxlsx.csv"
colunas_desejadas = [
    "UF",
    "MUNICIPIO",
    "INSTITUICAO",
    "SEGMENTO",
    "BAIRRO"
]
mapeamento = {
    "NOME INSTITUIÇÃO": "INSTITUICAO",
    "NOME_INSTITUICAO": "INSTITUICAO",
    "NOME INSTITUICAO": "INSTITUICAO",
    "NOME DA INSTITUIÇÃO": "INSTITUICAO",
    "NOME DA INSTITUICAO": "INSTITUICAO",
    "ESTADO": "UF",
    "MUNICÍPIO": "MUNICIPIO",
    "MUNICIPIO": "MUNICIPIO",
    "SEGMENTOS": "SEGMENTO",
    "BAIRRO": "BAIRRO"
}
```

```python
def normalizar_coluna(col):
    col = str(col).strip().upper()
    col = unicodedata.normalize("NFKD", col).encode("ASCII", "ignore").decode("utf-8")
    return col
```

```python
df = pd.read_excel(r"C:\Users\samuelbarroso\Downloads\Agencias\Agencias\200810AGENCIAS.xls",skiprows=8,header=1)
df.columns = df.columns.astype(str).str.strip()
df = df.loc[:, ~df.columns.duplicated()]
df.columns = [normalizar_coluna(c) for c in df.columns]
df = df.rename(columns=mapeamento)
df = df[colunas_desejadas]
df.head()
```

```python
files = sorted([f for f in os.listdir(folder) if f.endswith(('.xls'))])
```

```python
dataframes = []
for idx, file in enumerate(files):
    print(f"[{idx+1}/{len(files)}] Lendo {file}...")
    try:
        df = pd.read_excel(os.path.join(folder, file),skiprows=8,header=1)
        df.columns = df.columns.astype(str).str.strip()
        df = df.loc[:, ~df.columns.duplicated()]
        df.columns = [normalizar_coluna(c) for c in df.columns]
        df = df.rename(columns=mapeamento)
        df = df[colunas_desejadas]
        df = df.dropna(how="all").reset_index(drop=True)
        df['Arquivo_Origem'] = pd.to_datetime(file[:6], format="%Y%m")
        df['Quantidade'] = 1
        dataframes.append(df)
    except Exception as e:
        print(f"Erro ao ler {file}: {e}")
```

```python

if dataframes:
    print("Concatenando...")
    df_final = pd.concat(dataframes, ignore_index=True)
    print("Salvando em CSV...")
    df_final.to_csv(output_fileXLS, index=False, sep=';', encoding='utf-8-sig') # Using semicolon and utf-8-sig for better excel compatibility in BR
    print(f"Arquivo salvo com sucesso: {output_fileXLS}")
else:
    print("Nenhum dado encontrado para consolidar.")
```

```python
if dataframes:
    print("Concatenando...")
    df_final = pd.concat(dataframes, ignore_index=True)
    print("Salvando em CSV...")
    df_final.to_csv(output_fileXLSX, index=False, sep=';', encoding='utf-8-sig') # Using semicolon and utf-8-sig for better excel compatibility in BR
    print(f"Arquivo salvo com sucesso: {output_fileXLSX}")
else:
    print("Nenhum dado encontrado para consolidar.")
```
